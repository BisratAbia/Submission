```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
 
import matplotlib.ticker as tkr
%matplotlib inline
 
import glob
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import xml.etree.ElementTree as ET
import geopy.distance
from datetime import datetime
import sys

import pandas as pd
import numpy as np
from sklearn.cross_validation import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
from matplotlib import pyplot as plt 
%matplotlib inline 

from sklearn.ensemble import RandomForestRegressor,BaggingRegressor,GradientBoostingRegressor,AdaBoostRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split,cross_validate
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_log_error,mean_squared_error, r2_score,mean_absolute_error 
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score  
```

```python
path ='2017_sara'
allFiles = glob.glob(path + "/*.csv")
frame = pd.DataFrame()
col_names = ['rental_id','duration','bike_id','end_date','end_station_id','end_station_name','start_date',
             'start_station_id','start_station_name']
df_list = []

cols=list(range(9))
for file_ in allFiles:
    
    df = pd.read_csv(file_,names=col_names,usecols=cols,header=0, engine='python',
                     parse_dates=['start_date'],dayfirst=True,infer_datetime_format=True)
    df_list.append(df)
frame=pd.concat(df_list,ignore_index=True)
#drop missing values: negligible number of hire data has no duration/timestamps
frame = frame.dropna(axis=0, how='any')
frame = frame.reset_index(drop=True)
# frame['day_of_week'] = frame['start_date'].dt.dayofweek
# frame['hour'] = frame['start_date'].dt.hour
# frame['day'] = frame['start_date'].dt.day
# frame['month'] = frame['start_date'].dt.month
# frame['year'] = frame['start_date'].dt.year
# frame['quarter'] = frame['start_date'].dt.quarter
# frame['weekday'] = frame['start_date'].dt.weekday

frame.head(12)
```

```python
path ='weather/2017_weather_sara'
allFiles2 = glob.glob(path + "/*.csv")
frame2 = pd.DataFrame()
 
col_names2 = ['Date','rain_acc','air_temperature','Quality_flag','horizontal_wind_speed','relative_humidity','wind_direction']
df_list2 = []
cols=list(range(7))
for file2_ in allFiles2:   
    df2 = pd.read_csv(file2_,names=col_names2,usecols=cols,header=0, engine='python',
                     parse_dates=['Date'],dayfirst=True,infer_datetime_format=True)
    df_list2.append(df2)
    #df2['Date'] = df2['Date'].astype('datetime64[ns]')
    #df2['Date']= pd.to_datetime(df2['Date'].str.strip(),dayfirst=True)
frame2=pd.concat(df_list2,ignore_index=True)
frame2 = frame2.reset_index(drop=True)

frame2['Date'] = frame2['Date'].astype('datetime64[ns]')
#frame2.fillna(value=0.0, inplace=True)
frame2['rain_acc'] = frame2['rain_acc'].convert_objects(convert_numeric=True)
frame2.fillna(method='ffill', inplace=True)
frame2['air_temperature'] = frame2['air_temperature'].convert_objects(convert_numeric=True)
frame2['horizontal_wind_speed'] = frame2['horizontal_wind_speed'].convert_objects(convert_numeric=True)
frame2['relative_humidity'] = frame2['relative_humidity'].convert_objects(convert_numeric=True)
frame2['wind_direction'] = frame2['wind_direction'].convert_objects(convert_numeric=True)
frame2.head()
```

```python
frame2.air_temperature.isnull().values.any()
```

```python
rides_per_day = pd.DataFrame(frame.groupby('start_date').size().reset_index(name='rides'))

merged=  pd.merge(rides_per_day, frame2, left_on=['start_date'], right_on=['Date'])
merged = merged[['Date','rides','rain_acc','air_temperature','relative_humidity']]
merged['day_of_week'] = merged['Date'].dt.dayofweek
merged['hour'] = merged['Date'].dt.hour
merged['day'] = merged['Date'].dt.day
merged['month'] = merged['Date'].dt.month
#merged['year'] = merged['Date'].dt.year
#merged['quarter'] = merged['Date'].dt.quarter
merged['weekday'] = merged['Date'].dt.weekday
peak = pd.Series(merged['hour'])
merged['peak_hour'] = peak.isin([7, 8, 9, 16,17, 18, 19]).astype(int)

#test_df['peak_hour'] = peak.isin([7, 8, 9, 17, 18, 19]).astype(int)
merged.head(10)
```

```python
train_feature = merged.drop(['Date','rides'], axis = 1)
train_feature.head()
```

```python
enc = preprocessing.OneHotEncoder()
encoded = enc.fit_transform(train_feature[['day_of_week',  'month', 'day', 'hour', 'weekday', 'peak_hour']]).toarray()

#standardscaler numberic data
scaled = preprocessing.scale(train_feature[['air_temperature','relative_humidity']].values) 
 
trans_feature = np.concatenate([encoded, scaled], axis =1)

features = ['day_of_week', 'month', 'day', 'hour', 'weekday', 'peak_hour','air_temperature','relative_humidity']

print ("encoded categorical data",'\n', encoded, np.shape(encoded), '\n') 
print ("scaled numeric data", '\n', scaled, np.shape(scaled), '\n')
print ("transformed data", '\n', trans_feature, np.shape(trans_feature), '\n')
print ("selected features: ", '\n', features)
```

```python
target = merged['rides']
#split train/test
train_data, test_data, train_target, test_target = train_test_split (trans_feature, target, test_size = 0.1, random_state = 2000)

print ("train_data", train_data.shape,'\n', train_data, '\n')
print ('test_data', test_data.shape, '\n', test_data)
```
